```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()

url = 'https://finance.yahoo.com/quote/BTC-USD/news/'
driver.get(url)

time.sleep(5)

html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')

news_titles = soup.find_all('a', class_='subtle-link fin-size-small titles noUnderline yf-1i440lo')

if news_titles:
    for title_tag in news_titles:
        title = title_tag.text.strip()
        print(f"Título: {title}")
        print("-" * 30)
else:
    print("No se encontraron títulos.")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import datetime

now = datetime.datetime.now()

day = now.day

formatted_date = now.strftime("%a, %B %d, %Y").replace(' 0', ' ')

options = Options()
options.add_argument("--headless")
options.add_argument("--window-size=1920,1080")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=options)

url = 'https://finance.yahoo.com/quote/BTC-USD/news/'
driver.get(url)

time.sleep(5)

try:
    try:
        driver.switch_to.frame("render-target-default")
        accept_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Aceptar todo')]")
        accept_button.click()
        driver.switch_to.default_content()
        print("Ventana de cookies aceptada desde iframe.")
        time.sleep(3)
    except:
        accept_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Aceptar todo')]")
        accept_button.click()
        print("Ventana de cookies aceptada.")
        time.sleep(3)
except Exception as e:
    print("No se pudo aceptar la ventana de cookies. Error:", e)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

news_articles = soup.find_all('a', class_='subtle-link fin-size-small titles noUnderline yf-1i440lo')

if news_articles:
    for article in news_articles:
        title = article.text.strip()
        link = article['href']

        driver.get(link)
        time.sleep(3)

        full_article_html = driver.page_source
        full_article_soup = BeautifulSoup(full_article_html, 'html.parser')
        full_articles = full_article_soup.find('div', class_='body yf-1ir6o1g')

        if full_articles:
            date_element = full_article_soup.select_one('div.byline-attr-time-style time')

            if date_element:
                date = date_element.text.strip()
                today = date.split(" at ")[0]

                if today == formatted_date:
                    article_section_1 = full_articles.find('div', class_='atoms-wrapper')
                    text_section_1 = ""
                    if article_section_1:
                        paragraphs_1 = article_section_1.find_all('p', class_='yf-1090901')
                        text_section_1 = " ".join([p.text.strip() for p in paragraphs_1])

                    article_section_2 = full_articles.find('div', class_='read-more-wrapper')
                    text_section_2 = ""
                    if article_section_2:
                        paragraphs_2 = article_section_2.find_all('p', class_='yf-1090901')
                        text_section_2 = " ".join([p.text.strip() for p in paragraphs_2])

                    
                    full_article_text = text_section_1 + text_section_2

                    print(f"Título: {title}")
                    print(f"Link: {link}")
                    print(f"Date: {date}")
                    print(f"Text: {full_article_text}")
                    print("-" * 30)

        else:
            date = "Fecha no encontrada"

else:
    print("No se encontraron títulos.")

driver.quit()
```
